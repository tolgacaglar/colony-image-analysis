```python
import glob
import os
import shutil
```

```python
exp_name = "EQ59_Glu_03142021"
acq_name = "TileScan_48h-72h_xyzt"

exp_folder = os.path.join("E:", "Tolga", f"{exp_name}")
tilescan_folder = os.path.join(exp_folder, "TileScan")

acq_path_holder = os.path.join(tilescan_folder, f"{exp_name}_{acq_name}*tif")
from_files = glob.glob(acq_path_holder)
```

```python
# Move files from to its own base folder
base_folder = os.path.join(exp_folder, exp_name)
for fix in range(len(from_files)):
    srcpath = from_files[fix]
    fname = from_files[fix].split("\\")[-1]
    destpath = os.path.join(tilescan_folder, f"{exp_name}_{acq_name}", fname)
    if (fix%300 == 0):
        print(f"Moving file {fix}/{len(from_files)}")
    shutil.move(srcpath, destpath)
```
